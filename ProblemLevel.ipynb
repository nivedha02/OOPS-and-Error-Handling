1. Bug report class - easy
Create a class called BugReport with attributes: title, severity (low/medium/high), status (open by default). 
Add a method resolve() that changes status to "closed" and a method summary() that prints a one-line description.

bug = BugReport("Login fails on Safari", "high")
bug.summary()    # "Bug: Login fails on Safari | Severity: high | Status: open"
bug.resolve()
bug.summary()    # "Bug: Login fails on Safari | Severity: high | Status: closed"

```python
class BugReport:
    def __init__(self, title, severity, status="open"):
        self.title = title
        self.severity = severity
        self.status = status
    
    def resolve(self):
        self.status = "closed"
    
    def summary(self):
        print(f"Bug: {self.title} | Severity: {self.severity} | Status: {self.status}")

bug = BugReport("Login fails on Safari", "high")
bug.summary()    # "Bug: Login fails on Safari | Severity: high | Status: open"
bug.resolve()
bug.summary()    # "Bug: Login fails on Safari | Severity: high | Status: closed"
```

2. Test case class - easy
Create a TestCase class with name, expected, and actual attributes. Add a method passed() that returns True if expected equals actual, and a method result() that prints PASS or FAIL.

tc = TestCase("Login test", "Welcome, User!", "Welcome, User!")
tc.result()   # PASS: Login test
tc2 = TestCase("Price test", "99.00", "98.99")
tc2.result()  # FAIL: Price test

```python
class TestCase:
    def __init__(self,name,expected,actual):
        self.name=name
        self.expected=expected
        self.actual=actual
    def passed(self):
        return(self.expected==self.actual)
    def result(self):
        if self.passed():
            print(f"PASS: {self.name}")
        else:
            print(f"FAIL: {self.name}")

tc = TestCase("Login test", "Welcome, User!", "Welcome, User!")
tc.result()   # PASS: Login test
tc2 = TestCase("Price test", "99.00", "98.99")
tc2.result()  # FAIL: Price test
```

3. Test suite class - medium
Build a TestSuite class that holds a list of TestCase objects (from problem 2). 
Add methods: add(tc) to add a test, run() to print all results, and summary() to print total passed and failed count.

suite = TestSuite()
suite.add(TestCase("Login", "200", "200"))
suite.add(TestCase("Logout", "200", "404"))
suite.run()
suite.summary()  # Passed: 1 | Failed: 1

```python
class TestSuite:
    def __init__(self):
        self.list1=[]
        self.Passed=0
        self.Failed=0
    def add(self,tc):
        self.list1.append(tc)
        
    def run(self):
        for i in self.list1:
            if (i.passed()):
                self.Passed+=1
            else:
                self.Failed+=1

    def summary(self):
        print(f"Passed: {self.Passed} | Failed: {self.Failed}")

suite = TestSuite()
suite.add(TestCase("Login", "200", "200"))
suite.add(TestCase("Logout", "200", "404"))
suite.run()
suite.summary()  # Passed: 1 | Failed: 1
```

6. Custom exception - medium
Create a custom exception InvalidSeverityError. Modify BugReport.__init__ to raise this exception if severity is not low/medium/high.

try:
    bug = BugReport("Login broken", "critical")
except InvalidSeverityError as e:
    print(e)  # Invalid severity: critical. Use low, medium, or high.

```python
class InvalidSeverityError(Exception):
    pass
class BugReport:
    def __init__(self, title, severity, status="open"):
        self.title = title
        self.severity = severity
        self.status = status
        if self.severity in ("low", "medium", "high"):
            print(f"Severity:{self.severity}")
        raise InvalidSeverityError(f"Invalid severity: {self.severity}. Use low, medium, or high.")
        
try:
    bug = BugReport("Login broken", "critical")
except InvalidSeverityError as e:
    print(e)  # Invalid severity: critical. Use low, medium, or high.
```

7. Try / except / finally - medium
Write a function load_test_data(filename) that opens a JSON file and returns its contents. Handle FileNotFoundError and json.JSONDecodeError separately. Use finally to always print "Done loading."

load_test_data("tests.json")   # works if file exists

```python
def load_test_data(filename):
    try:
        with open (filename,"r") as file:
            content=file.read()
            print(content)
    except FileNotFoundError:
        print(f"File not found: missing.json")
    finally:
        print("Done Loading")

load_test_data("Qtest.json")   # works if file exists
load_test_data("missing.json") # File not found: missing.json
                               # Done loading.
```

8. Save bugs to a file - medium
Add a method to_dict() to BugReport. Then write a function save_bugs(bugs, filename) that saves a list of bug dicts to a JSON file.

bugs = [BugReport("Login broken", "high"), BugReport("UI glitch", "low")]
save_bugs(bugs, "bugs.json")
# bugs.json → [{"title": "Login broken", "severity": "high", "status": "open"}, ...]

```python
import json
class BugReport:
    def __init__(self, title, severity, status="open"):
        self.title = title
        self.severity = severity
        self.status = status
    def to_dict(self):
        return {
            "title": self.title,
            "severity": self.severity,
            "status": self.status
        }
def save_bugs(bugs,filename):
    bug_data=[bug.to_dict() for bug in bugs]
    with open(filename,'w') as file:
        json.dump(bug_data,file)
        #json.dump(bug_data,file,indent=4) #indent=4 gives indent=4 means each nested level is indented by 4 spaces.
        #It makes the JSON easier for humans to read. It does not change the actual data.


bugs = [BugReport("Login broken", "high"), BugReport("UI glitch", "low")]
save_bugs(bugs, "Rbugs.json")
# bugs.json → [{"title": "Login broken", "severity": "high", "status": "open"}, ...]
```

9. API response validator class - hard
Build an APIValidator class that takes a response dict with status_code and body. Add methods: is_success() returns True if status 200–299, has_key(key) returns True if key exists in body, validate(required_keys) prints pass/fail for each key.

resp = {"status_code": 200, "body": {"user_id": 1, "token": "abc"}}
v = APIValidator(resp)
print(v.is_success())        # True
v.validate(["user_id", "email"])
# PASS: user_id
# FAIL: email — key missing

```python
class APIValidator:
    def __init__(self,response):
        self.status_code=response["status_code"]
        self.body=response["body"]
    def is_success(self):
        return 200<=self.status_code<=299
    def  has_key(self,key):
        return key in self.body
    def validate(self,required_keys):
        for key in required_keys:
            if self.has_key(key):
                print(f"Pass: {key}")
            else:
                print(f"Fail: {key}")
resp = {"status_code": 200, "body": {"user_id": 1, "token": "abc"}}
v = APIValidator(resp)
print(v.is_success())        # True
v.validate(["user_id", "email"])       
```

10. Mini test runner - hard
Combine everything. Build a TestRunner class that accepts a list of functions to run as tests. Each function should return True (pass) or raise an exception (fail). The runner catches exceptions, logs the error message, and prints a final summary.

def test_addition():
    assert 1 + 1 == 2
    return True

def test_login():
    raise Exception("Login API timeout")

runner = TestRunner([test_addition, test_login])
runner.run()
# PASS: test_addition
# FAIL: test_login — Login API timeout
# Summary: 1 passed, 1 failed

```python
class TestRunner:
    def __init__(self, tests):
        self.tests = tests

    def run(self):
        passed = 0
        failed = 0

        for test in self.tests:
            try:
                test()
                print(f"PASS: {test.__name__}")
                passed += 1

            except Exception as e:
                print(f"FAIL: {test.__name__} — {e}")
                failed += 1

        print(f"Summary: {passed} passed, {failed} failed")


def test_addition():
    assert 1 + 1 == 2
    return True


def test_login():
    raise Exception("Login API timeout")

runner = TestRunner([test_addition, test_login])
runner.run()
# PASS: test_addition
# FAIL: test_login — Login API timeout
# Summary: 1 passed, 1 failed
```
